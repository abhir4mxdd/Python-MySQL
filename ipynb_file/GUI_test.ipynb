```python
def dev_add() :
    updev = db.cursor()
    did = dev_id.get()
    dip = ip.get()
    port = portable.get()
    ops = os.get()
    pr = int(price.get())
    if did == '' or dip == '' or port == '' or ops == '' or pr == None:
        message.set('Fill in the Required Fields !')
    else : 
        query = ("INSERT INTO DEVICE(device_id, IP, portability, price, os)"
           "VALUES (%s, %s, %s, %s, %s)")
        data = (did , dip , port , pr , ops)
        updev.execute(query , data)
        db.commit()
        #db.rollback()
        message.set("Stored successfully  !")
```
